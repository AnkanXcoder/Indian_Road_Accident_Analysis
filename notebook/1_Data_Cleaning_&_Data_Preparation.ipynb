# **Data Type Validation & correction**

```python
df.info()
```

```python
df['accident_id'] = df['accident_id'].astype(np.int16)
df['latitude'] = df['latitude'].astype(np.float32)
df['longitude'] = df['longitude'].astype(np.float32)
df['hour'] = df['hour'].astype(np.int8)
df['is_weekend'] = df['is_weekend'].astype(np.int8)
df['lanes'] = df['lanes'].astype(np.int8)
df['traffic_signal'] = df['traffic_signal'].astype(np.int8)
df['temperature'] = df['temperature'].astype(np.int8)
df['vehicles_involved'] = df['vehicles_involved'].astype(np.int8)
df['is_peak_hour'] = df['is_peak_hour'].astype(np.int8)
df['risk_score'] = df['risk_score'].astype(np.float32)
df['date'] = pd.to_datetime(df['date'])
```

```python
df.info()
```

# **Feature Engineering**

1. **Severity Index**

```python
df['severity_index'] = (
    df['vehicles_involved'] *
    df['casualties']
)
```

2. **Risk Quartile**

```python
df['risk_quartile'] = pd.qcut(
    df['risk_score'],
    q=4,
    labels=['Q1', 'Q2', 'Q3', 'Q4']
)
```

3. **Weekend Label Change**

```python
df['is_weekend'] = df['is_weekend'].replace({
    0:'Weekday',
    1:'Weekend'
})
```

4. **Peak Hour Label**

```python
df['is_peak_hour'] = df['is_peak_hour'].replace({
    0:'Non Peak Hour',
    1:'Peak Hour'
})
```
